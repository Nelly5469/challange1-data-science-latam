### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
colores = px.colors.sequential.Magma_r
print("Colores en hexadecimal:", colores)

# Muestra visual de los colores
plt.figure(figsize=(8, 2))
for i, color in enumerate(colores):
    plt.fill_between([i, i+1], 0, 1, color=color)
plt.xlim(0, len(colores))
plt.yticks([])
plt.xticks(range(len(colores)), colores, rotation=90, fontsize=8)
plt.title("Paleta Magma_r")
plt.show()
```

# 1. Análisis de facturación



## Ingreso total por tienda

```python
ingresos_tienda1 = tienda['Precio'].sum()
ingresos_tienda2 = tienda2['Precio'].sum()
ingresos_tienda3 = tienda3['Precio'].sum()
ingresos_tienda4 = tienda4['Precio'].sum()
print(f"Ingresos Tienda 1: ${ingresos_tienda1:,.2f}")
print(f"Ingresos Tienda 2: ${ingresos_tienda2:,.2f}")
print(f"Ingresos Tienda 3: ${ingresos_tienda3:,.2f}")
print(f"Ingresos Tienda 4: ${ingresos_tienda4:,.2f}")
```

```python
ingresos_tiendas = pd.Series([ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4],
                            index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
pie_chart_ingresos = px.pie(ingresos_tiendas, values=ingresos_tiendas.values, names=ingresos_tiendas.index,
                            title='Ingresos por Tienda', 
                            color_discrete_sequence=px.colors.sequential.Magma_r,
                            # En gráficos de pastel, no funciona color_discrete_sequence para valores continuos, solo en px.bar, px.scatter, etc.
                            #color=ingresos_tiendas.values,  # Asocia el color al valor continuo
                            #color_continuous_scale=px.colors.sequential.Viridis  # Escala de colores continua
                            )
pie_chart_ingresos.show()
```

# 2. Ventas por categoría

## Tienda 1

```python
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en Tienda 1:")
print(ventas_categoria_tienda1)
```

## Tienda 2

```python
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en Tienda 2:")
print(ventas_categoria_tienda2)
```

## Tienda 3

```python
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en Tienda 3:")
print(ventas_categoria_tienda3)
```

## Tienda 4

```python
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
print("Ventas por categoría en Tienda 4:")
print(ventas_categoria_tienda4)
```

## Visualización

### Formas de acomodar el dataframe antes del gráfico

Ver más info en: 

    [Gráficos de barras en Plotly](https://plotly.com/python/bar-charts/)

    [Argumentos para gráficos Plotly](https://plotly.com/python/px-arguments/)

```python
ventas_por_categoria_tiendas = pd.DataFrame({
    'Tienda 1': ventas_categoria_tienda1,
    'Tienda 2': ventas_categoria_tienda2,
    'Tienda 3': ventas_categoria_tienda3,
    'Tienda 4': ventas_categoria_tienda4})

ventas_por_categoria_tiendas.head()
```

```python
wide_ventas_categoria_tiendas = ventas_por_categoria_tiendas.transpose()
wide_ventas_categoria_tiendas.head()
```

```python
print(wide_ventas_categoria_tiendas.index)
print(wide_ventas_categoria_tiendas.columns)
wide_ventas_categoria_tiendas.index.name = 'Tienda'
print(wide_ventas_categoria_tiendas.index)
```

### Gráfico barras apiladas

Con wide format

```python
ventas_por_categoria_tiendas_plot = px.bar(wide_ventas_categoria_tiendas,
                                           #x = wide_ventas_categoria_tiendas.index,
                                           #y = wide_ventas_categoria_tiendas.columns,
                                           title='Ventas por Categoría y Tienda',
                                           labels={'value':'Ventas Totales', 'Tienda':'Tiendas', 'variable':'Categoría del Producto'},
                                           #color_discrete_sequence=px.colors.qualitative.Pastel,
                                           color_discrete_sequence=px.colors.sequential.Magma_r,
                                           x = wide_ventas_categoria_tiendas.columns,
                                            y = wide_ventas_categoria_tiendas.index,
                                           orientation='h'
                                           )
ventas_por_categoria_tiendas_plot.show()
```

# 3. Calificación promedio de la tienda


## Tienda 1

```python
cal_prom_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio Tienda 1: {cal_prom_tienda1}")
```

## Tienda 2

```python
cal_prom_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio Tienda 2: {cal_prom_tienda2}")
```

## Tienda 3

```python
cal_prom_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio Tienda 3: {cal_prom_tienda3}")
```

## Tienda 4

```python
cal_prom_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio Tienda 4: {cal_prom_tienda4}")
```

## Visualización

```python
calificaciones_tiendas = pd.Series([cal_prom_tienda1, cal_prom_tienda2, cal_prom_tienda3, cal_prom_tienda4],
                                  index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
print("Calificaciones Promedio por Tienda:")
print(calificaciones_tiendas)
```

```python
calificaciones_tiendas_scatter_plot = px.scatter(calificaciones_tiendas, 
                                        x=calificaciones_tiendas.index, 
                                        y=calificaciones_tiendas.values,
                                        title='Calificación Promedio por Tienda',
                                        labels={'x':'Tiendas', 'y':'Calificación Promedio'},
                                        color_discrete_sequence=['#cd4071'],
                                        size=calificaciones_tiendas.values)
calificaciones_tiendas_scatter_plot.show()
```
